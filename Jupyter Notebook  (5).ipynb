```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
soup = BeautifulSoup(r.content, 'lxml')
table = soup.find_all('table')[0]

df = pd.read_html(str(table))

neighborhood=pd.DataFrame(df[0])
```

```python
postcode = []
borough = []
neighborhood = []
table_post = soup.find('table')
fields = table_post.find_all('td')
```

```python
for i in range(0, len(fields), 3):
    postcode.append(fields[i].text.strip())
    borough.append(fields[i+1].text.strip())
    neighborhood.append(fields[i+2].text.strip())
```

```python
df2 = pd.DataFrame(data=[postcode, borough, neighborhood]).transpose()
df2.columns = ['Postal Code', 'Borough', 'Neighborhood']
df2.head(10)
```

Below I removed all rows which have no assigned neighborhood

```python
df3 = df2[df2['Borough'] !='Not assigned']
df3.head()
```

Below I have grouped all my entries by their postcode

```python
df3.groupby("Postal Code")
```

```python
other_path = "https://cocl.us/Geospatial_data"
df11 = pd.read_csv(other_path, header=0)
df11.head()
```

Below I have merged my dataframes on the column 'Postal Code'

```python
df12 = pd.merge(df3,df11, on='Postal Code')
df12.head()
```

```python
!conda install -c conda-forge folium=0.5.0 --yes 
import folium 
```

```python
address = 'Toronto, CA'

geolocator = Nominatim(user_agent="CA_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of toronto City are {}, {}.'.format(latitude, longitude))
```

Below is a map of toronto with all the neighborhoods superimposed on top

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

for lat, lng, borough, neighborhood in zip(df12['Latitude'], df12['Longitude'], df12['Borough'], df12['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

In the above map we can see that the neighborhoods cluseter heavily in the southern center of Toronto and become more spread out the further you go.
